```python

df_voitures=pd.read_csv('voitures.csv',names=('immat','marque','km','date_cir'),sep=';')
```

```python
df_voitures
```

```python
X_=df_voitures.copy()
```

```python
X_.set_index('immat', inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
ohe.fit(X_[['marque']])
```

```python
ohe.categories_[0]
```

```python
marque_encoded=ohe.transform(X_[['marque']])
```

```python
marque_encoded=marque_encoded.toarray()
```

```python
marque_encoded
```

```python
X_marque_encoded=pd.DataFrame(marque_encoded,
                                    index=X_.index,
                                    columns=ohe.categories_[0])
```

```python
X_marque_encoded
```

```python
X_=pd.concat([X_,X_marque_encoded], axis=1)
```

```python
X_
```
